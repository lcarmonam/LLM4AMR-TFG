```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dynamic dataset 
c = pd.read_csv('./df_14days_dynamic_23072024_allFeatures.csv')

# Filter out invalid patients
data_14_days = c[c.individualMRGerm != 666]
data_14_days = data_14_days.groupby("Admissiondboid").sum().reset_index()

# Create binary AMR label: 1 if patient developed AMR, else 0
data_14_days['individualMRGerm_binary'] = data_14_days['individualMRGerm'].apply(lambda x: 1 if x > 0 else 0)

# Define the list of antibiotic family columns 
antibiotic_cols = [
    'CAR', 'PEN', 'PAP', 'AMG', 'CF1', 'CF2', 'CF3', 'CF4', 'GLI', 'QUI', 'SUL',
    'LIN', 'MAC', 'OXA', 'POL', 'LIP', 'NTI', 'TTC', 'ATF', 'MON', 'GCC', 'ATI',
    'ATP', 'OTR', 'Falta'
]

# Determine whether each patient received each antibiotic 
abx_presence = data_14_days[['Admissiondboid', 'individualMRGerm_binary'] + antibiotic_cols].copy()
for col in antibiotic_cols:
    abx_presence[col] = abx_presence[col].apply(lambda x: 1 if x > 0 else 0)

# Compute percentage of patients treated with each antibiotic 
abx_pct = abx_presence.groupby('individualMRGerm_binary')[antibiotic_cols].mean().T * 100
abx_pct.columns = ['Non-AMR', 'AMR']
abx_pct_sorted = abx_pct.sort_values(by='AMR', ascending=False)

# Convert to long format for seaborn visualization 
abx_pct_long = abx_pct_sorted.reset_index().melt(id_vars='index', value_vars=['Non-AMR', 'AMR'])
abx_pct_long.columns = ['Antibiotic', 'Group', 'Percentage']

# Plot grouped bar chart (antibiotic vs % of patients treated) 
plt.figure(figsize=(14, 6))
sns.barplot(
    data=abx_pct_long,
    x='Antibiotic',
    y='Percentage',
    hue='Group',
    palette={'Non-AMR':  '#4E79A7', 'AMR':  '#E15759'},
    dodge=True
)


plt.ylabel(" ")
plt.xlabel(" ")
plt.xticks(rotation=45, ha='right', fontsize=18)
plt.yticks(fontsize=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(fontsize=17)
plt.tight_layout()

# Save the figure
plt.savefig("eda_antibiotics_patient_proportion.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
c = pd.read_csv('./df_14days_dynamic_23072024_allFeatures.csv')
c = c[c['individualMRGerm'] != 666].copy()

# Assign AMR group
amr_labels = c.groupby('Admissiondboid')['individualMRGerm'].sum().reset_index()
amr_labels['Group'] = amr_labels['individualMRGerm'].apply(lambda x: 'AMR' if x > 0 else 'Non-AMR')
c = c.merge(amr_labels[['Admissiondboid', 'Group']], on='Admissiondboid', how='left')

# Define all variables (interventions + organ failures)
all_vars = {
    # Clinical interventions
    'nutr_art': 'Artificial nutrition',
    'sedacion': 'Sedation',
    'relajacion': 'Relaxation',
    'insulina': 'Insulin',
    'cambio_postural': 'Postural changes',
    
    # Organ failures
    'fallo_hepatico': 'Liver failure',
    'fallo_renal': 'Renal failure',
    'fallo_coagulacion': 'Coagulation failure',
    'fallo_hemodinamico': 'Hemodynamic failure',
    'fallo_respiratorio': 'Respiratory failure',
    'fallo_multiorganico': 'Multiorgan failure'
}

# Aggregate to patient level (1 if occurred at any point)
all_by_patient = c.groupby(['Admissiondboid', 'Group'])[list(all_vars.keys())].max().reset_index()

# Compute percentages
all_percentages = (
    all_by_patient.groupby('Group')[list(all_vars.keys())]
    .mean().T * 100
).round(1)

# Rename rows
all_percentages.index = [all_vars[var] for var in all_percentages.index]
all_percentages = all_percentages.reset_index().melt(id_vars='index', var_name='Group', value_name='Percentage')
all_percentages.rename(columns={'index': 'Variable'}, inplace=True)

# Plot single figure
plt.figure(figsize=(12, 7))
sns.barplot(
    data=all_percentages,
    x='Variable',
    y='Percentage',
    hue='Group',
    palette={'Non-AMR': '#4E79A7', 'AMR': '#E15759'}
)


plt.ylabel(' ')
plt.xlabel('')
plt.xticks(rotation=45, ha='right', fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=17)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('eda_combined_interventions_and_failures.png', dpi=300)
plt.show()

# summary table
summary_table = all_percentages.pivot(index='Variable', columns='Group', values='Percentage')
print("\nPercentage of patients affected by each clinical variable (per group):")
display(summary_table)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('./df_14days_dynamic_23072024_allFeatures.csv')
df = df[df['individualMRGerm'] != 666].copy()  

# Label patients as AMR or Non-AMR based on whether they ever had a resistant germ
amr_labels = df.groupby('Admissiondboid')['individualMRGerm'].sum().reset_index()
amr_labels['Group'] = amr_labels['individualMRGerm'].apply(lambda x: 'AMR' if x > 0 else 'Non-AMR')
df = df.merge(amr_labels[['Admissiondboid', 'Group']], on='Admissiondboid', how='left')

# Create a third group: all patients 
df_all = df.copy()
df_all['Group'] = 'All patients'
df_combined = pd.concat([df, df_all])

# Define hospital day 
df_combined['Day'] = df_combined['dayToDone'] + 1

# Calculate % of patients with mechanical ventilation (hoursVM > 0), per day and group
summary = []

for group in ['AMR', 'Non-AMR']:
    group_df = df_combined[df_combined['Group'] == group]
    for day in sorted(group_df['Day'].unique()):
        day_df = group_df[group_df['Day'] == day]
        total = len(day_df)
        on_vm = (day_df['hoursVM'] > 0).sum()
        percentage = (on_vm / total) * 100 if total > 0 else 0
        summary.append({
            'Group': group,
            'Day': day,
            'Percent_on_VM': round(percentage, 2),
            'Total_patients': total,
            'Patients_on_VM': on_vm
        })

df_percent_vm = pd.DataFrame(summary)

# Display final table
print(df_percent_vm)
```

### Microbiological culture data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('./df_14days_dynamic_23072024_allFeatures.csv')
df = df[df['individualMRGerm'] != 666].copy()

# Define culture-related binary variables
culture_vars = {
    'pc_acinet': 'Acinetobacter$_{\_pc}$',
    'pc_pseud': 'Pseudomonas$_{\_pc}$',
    'pc_enterobac': 'Enterobacter$_{\_pc}$',
    'pc_enteroc': 'Enterococcus$_{\_pc}$',
    'pc_staph': 'Staphylococcus aureus$_{\_pc}$',
    'pc_no_germ': 'Others$_{\_pc}$'
}

# Create AMR group label
amr_labels = df.groupby('Admissiondboid')['individualMRGerm'].sum().reset_index()
amr_labels['Group'] = amr_labels['individualMRGerm'].apply(lambda x: 'AMR' if x > 0 else 'Non-AMR')
df = df.merge(amr_labels[['Admissiondboid', 'Group']], on='Admissiondboid', how='left')

# Group by patient and take max per patient (if any detection over 14 days)
df_patient = df.groupby(['Admissiondboid', 'Group'])[list(culture_vars.keys())].max().reset_index()

# Compute proportions per group (only AMR and Non-AMR)
results = []
for col, label in culture_vars.items():
    for group in ['AMR', 'Non-AMR']:
        count = df_patient[df_patient['Group'] == group][col].sum()
        total = len(df_patient[df_patient['Group'] == group])
        results.append({
            'Organism': label,
            'Group': group,
            'Percentage': 100 * count / total
        })

df_results = pd.DataFrame(results)

# Plot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    data=df_results,
    x='Organism',
    y='Percentage',
    hue='Group',
    palette={'AMR': '#E15759', 'Non-AMR': '#4E79A7'}
)
barplot.legend(title=None)

# Annotate each bar with its value
for p in barplot.patches:
    height = p.get_height()
    if height > 0:
        barplot.annotate(f'{height:.1f}%',
                         (p.get_x() + p.get_width() / 2., height),
                         ha='center', va='bottom', fontsize=9, rotation=0)

plt.ylabel(' ')
plt.xlabel(' ')
plt.xticks(rotation=45, ha='right', fontsize=17)
plt.yticks(fontsize=17)
plt.legend(fontsize=17)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('prior_cultures_by_group.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
c = pd.read_csv('./df_14days_dynamic_23072024_allFeatures.csv')

# Filter out discharged patients and aggregate data at the patient level 
data_14_days = c[c['individualMRGerm'] != 666]
data_14_days = data_14_days.groupby("Admissiondboid").sum().reset_index()

# Create binary AMR label 
data_14_days['individualMRGerm_binary'] = data_14_days['individualMRGerm'].apply(lambda x: 1 if x > 0 else 0)

# Define visual order of antibiotics 
antibiotic_order_n = [
    'CAR', 'PAP', 'QUI', 'ATF', 'GLI', 'OXA', 'CF3', 'Falta', 'AMG',
    'PEN', 'SUL', 'CF4', 'NTI', 'LIN', 'CF1', 'MAC', 'POL', 'MON', 'LIP',
    'GCC', 'ATI', 'OTR', 'TTC', 'CF2', 'ATP'
]

# Build column names for neighbor exposure 
neighbor_abx_cols = [f'neighbor_{abx}' for abx in antibiotic_order_n]

# Create a reduced DataFrame and binarize exposure values
abx_neighbor = data_14_days[['Admissiondboid', 'individualMRGerm_binary'] + neighbor_abx_cols].copy()
for col in neighbor_abx_cols:
    abx_neighbor[col] = abx_neighbor[col].apply(lambda x: 1 if x > 0 else 0)

# Calculate percentage of patients exposed to each antibiotic, grouped by AMR status
abx_pct = abx_neighbor.groupby('individualMRGerm_binary')[neighbor_abx_cols].mean().T * 100

# Rename columns for clarity 
col_map = {0: 'Non-AMR', 1: 'AMR'}
abx_pct.columns = [col_map.get(col, str(col)) for col in abx_pct.columns]

# Add clean antibiotic names without the 'neighbor_' prefix and ensure desired display order
abx_pct['Antibiotic'] = [col.replace('neighbor_', '') for col in abx_pct.index]
abx_pct['Antibiotic'] = pd.Categorical(abx_pct['Antibiotic'], categories=antibiotic_order_n, ordered=True)
abx_pct_sorted = abx_pct.sort_values('Antibiotic')

# Convert to long format for plotting
abx_pct_long = abx_pct_sorted.drop(columns=['Antibiotic']).reset_index().melt(id_vars='index', value_vars=['Non-AMR', 'AMR'])
abx_pct_long.columns = ['Antibiotic', 'Group', 'Percentage']

# Add subscript '_n' to indicate neighbor exposure in the label
abx_pct_long['Antibiotic'] = abx_pct_long['Antibiotic'].str.replace('neighbor_', '', regex=False)
abx_pct_long['Antibiotic'] = abx_pct_long['Antibiotic'].apply(lambda x: f"{x}_n")

# Plot barplot comparing neighbor antibiotic exposure by group
plt.figure(figsize=(14, 6))
sns.barplot(
    data=abx_pct_long,
    x='Antibiotic',
    y='Percentage',
    hue='Group',
    palette={'Non-AMR': '#4E79A7', 'AMR': '#E15759'},
    dodge=True
)

# Styling
plt.ylabel('')
plt.xlabel('')
plt.xticks(rotation=45, ha='right', fontsize=18)
plt.yticks(fontsize=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(fontsize=17)
plt.tight_layout()

# Save and show figure
plt.savefig("eda_antibiotics_copatients.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out discharged patients 
daily_data = c[c['individualMRGerm'] != 666].copy()

# Create binary AMR label
daily_data['individualMRGerm_binary'] = daily_data['individualMRGerm'].apply(lambda x: 1 if x > 0 else 0)

# Define list of antibiotic codes 
antibiotic_cols = ['CAR', 'PAP', 'QUI', 'ATF', 'GLI', 'OXA']
neighbor_abx_cols = [f'neighbor_{abx}' for abx in antibiotic_cols]
abx_labels = [f"{abx}_n" for abx in antibiotic_cols]  # Label with "_n" to denote neighbor exposure

# Binarize exposure values per day
for col in neighbor_abx_cols:
    daily_data[col] = daily_data[col].apply(lambda x: 1 if x > 0 else 0)

# Generate one heatmap for each group
for group_value, group_name in [(1, "AMR"), (0, "Non-AMR")]:
    # Filter data for current group
    group_data = daily_data[daily_data['individualMRGerm_binary'] == group_value]

    # Group by day and calculate percentage of patients exposed to each antibiotic
    daily_exposure = group_data.groupby('dayToDone')[neighbor_abx_cols].mean().T * 100
    daily_exposure.index = abx_labels  
    daily_exposure = daily_exposure.loc[abx_labels] 

    # Plot heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(
        daily_exposure,
        annot=False,
        cmap="YlGnBu",
        linewidths=0.3,
        xticklabels=range(1, 15),  
        yticklabels=True
    )

    # Format axes and save figure
    cbar = plt.gcf().axes[-1]
    cbar.tick_params(labelsize=25)
    plt.xlabel(" ")
    plt.ylabel(" ")
    plt.xticks(fontsize=25)
    plt.yticks(fontsize=25, rotation=0)
    plt.tight_layout()
    plt.savefig(f"heatmap_neighbors_by_day_{group_name.lower()}.png", dpi=300)
    plt.show()

```
